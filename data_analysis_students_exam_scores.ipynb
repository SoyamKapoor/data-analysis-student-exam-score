## **Libraries Required**  

Ensure that the following libraries are installed before running the analysis:  


```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **1. Importing and Loading the Dataset**  
The dataset `Original_data.csv` is imported into a Pandas DataFrame.  
https://www.kaggle.com/datasets/desalegngeb/students-exam-scores

```python
# Importing the dataset
df=pd.read_csv("Original_data.csv")
```

### **Displaying the First Few Rows of the Dataset**

To get an initial overview of the dataset, we use the `.head()` function, which returns the first five rows by default.

```python
# Checking the first few rows of the dataset
print(df.head())
```

### **Removing Rows with Missing Values**

To clean the dataset by removing any rows that contain missing values (NaN), we use the `.dropna()` function.

```python
# Remove rows with any missing values (NaN) to clean the dataset
df_cleaned_file=df.dropna()
```

### **Identifying Numerical Columns in the Dataset**

To filter only the numerical columns from the cleaned dataset, we use the `.select_dtypes()` function with `np.number` as the parameter.

```python
# Identify the numerical columns in the dataset
df_numerical_only=df_cleaned_file.select_dtypes(include=[np.number])
```

### **Saving the Cleaned Dataset to a CSV File**

To save the cleaned numerical dataset as a new CSV file, we use the `.to_csv()` function.

```python
# Save the cleaned dataset to a new csv file
df_numerical_only.to_csv("cleaned_Original_data.csv",index=False)
```

### **Summary Statistics of the Numerical Columns**

To calculate key summary statistics such as mean, mode, and standard deviation for the numerical columns, we use the following functions:

```python
# Summary statistics of the numerical columns
mean_values = df_numerical_only.mean()
mode_values = df_numerical_only.mode().iloc[0]  # First mode value
std_dev_values = df_numerical_only.std()
```

### **Displaying the Summary Statistics**

To print the calculated summary statistics (mean, mode, and standard deviation) for the numerical columns, we use the following code:

```python
# Display the summary statistics
print(f"Mean Values:\n\n{mean_values}")
print(f"\nMode Values:\n\n{mode_values}")
print(f"\nStandard Deviation Values:\n\n{std_dev_values}")
```

### **Histogram of the Numerical Columns**

To visualize the distribution of numerical data, we use a histogram:

```python
# Histogram of the numerical columns
plt.hist(df_numerical_only, bins=5, alpha=1, label=df_numerical_only.columns)  
plt.title('Histogram of Numerical Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()  
```

## **Visualizing Normal Distribution Curves for Numerical Columns**

To analyze the distribution of numerical data, we overlay normal distribution curves onto the dataset. This helps in understanding how closely the data follows a normal distribution.

### **Code Implementation:**

```python
# Normal distribution curves for the numerical columns
x_range = np.linspace(df_numerical_only.values.min(), df_numerical_only.values.max(), 100)

for col in df_numerical_only.columns:
    mean = mean_values[col]
    std_dev = std_dev_values[col]
    normal_curve = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-((x_range - mean) ** 2) / (2 * std_dev ** 2))
    plt.plot(x_range, normal_curve, label=f'Normal Curve ({col})')

# Final plot settings
plt.title('Histogram with Normal Distribution Curves for Numerical Variables')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

## **Interpretation of the Results**

To analyze the numerical columns, we interpret key statistical measures such as mean, mode, and standard deviation.

### **Code Implementation:**

```python
# Interpretation of the results
print("Interpretation of Results:")
for col in df_numerical_only.columns:
    print(f"\nColumn: {col}")
    print(f"Mean ({col}): {mean_values[col]:.2f} → This represents the average value.")
    print(f"Mode ({col}): {mode_values[col]} → The most frequently occurring value in the dataset.")
    print(f"Standard Deviation ({col}): {std_dev_values[col]:.2f} → A higher value means more variation in data.")
```
